```python
import matplotlib.pyplot as plt
import os

color_list = ['g-','r-','b-','y-','k-']
def draw_from_file(filename, color_ids=0, label=None):
    x = []
    y = []
    step = 0
    with open(filename, 'r') as f:
        for i in f.readlines():
            if i.startswith('tensor'):
                loss = float(i.split(',')[0][len('tensor('):])
                step += 1
                if step > 1000:
                    x.append(step)
                    y.append(loss)
                
    
    plt.plot(x, y, color_list[color_ids], label=label)
if __name__ == '__main__':
    MoE_bias_raw_table_training_log_path = '/data1/fch123/UDT-QA/condenser/model_MoE_bias_raw_table/training.log'
    MoE_position_raw_table_training_log_path = '/data1/fch123/UDT-QA/condenser/model_MoE_position_raw_table/training.log'
    draw_from_file(MoE_bias_raw_table_training_log_path, 0, 'bias_raw_table')
    draw_from_file(MoE_position_raw_table_training_log_path, 1, 'position_raw_table')

    plt.legend()
    plt.show()
```
